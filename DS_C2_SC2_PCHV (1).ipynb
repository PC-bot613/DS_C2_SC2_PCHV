```python
import pandas as pd
df = pd.read_csv('vaccinations.csv')
```

```python
df.head()
```

## Estructura y tipos de datos de cada columna

```python
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

## Cantidad de vacunas aplicadas de cada compañía

```python
vacunas_por_compania = (
    df.groupby('vaccines', as_index=False)['total_vaccinations']
    .sum()
)

vacunas_por_compania
```

## Cantidad de vacunas aplicadas en todo el mundo.

```python
total_vacunas_mundo = df['total_vaccinations'].sum()
total_vacunas_mundo
```

## Promedio de vacunas aplicadas por país

```python
promedio_por_pais = (
    df.groupby('country', as_index=False)['total_vaccinations']
    .mean()
)

promedio_por_pais
```

## Cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
fecha_objetivo = pd.to_datetime('2021-01-29')

vacunas_290121 = (
    df[df['date'] == fecha_objetivo]['total_vaccinations']
    .sum()
)

vacunas_290121
```

## Dataframe conDiferencias

```python
conDiferencias = df.copy()

conDiferencias['diferencias'] = (
    conDiferencias['daily_vaccinations'] -
    conDiferencias['daily_vaccinations_raw']
)

conDiferencias.head()
```

## Periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
fecha_min = df['date'].min()
fecha_max = df['date'].max()

diferencia_tiempo = fecha_max - fecha_min
diferencia_tiempo
```

##Dataframe conCantidad

```python
conCantidad = df.copy()

conCantidad['canVac'] = conCantidad['vaccines'].str.split(',').apply(len)

conCantidad.head()
```

##Dataframe antes20

```python
fecha_limite = pd.to_datetime('2020-12-20')

antes20 = df[df['date'] < fecha_limite]

antes20.head()
```

##Dataframe Pfizer

```python
pfizer = df[df['vaccines'].str.contains('Pfizer', na=False)]

pfizer.head()
```

##Archivo de Excel resultadosReto.xlsx

```python
conDiferencias = df.copy()

conDiferencias['diferencias'] = (
    conDiferencias['daily_vaccinations'] -
    conDiferencias['daily_vaccinations_raw']
)

conDiferencias.head()

periodo_tiempo = df['date'].max() - df['date'].min()
periodo_tiempo

conCantidad = df.copy()

conCantidad['canVac'] = conCantidad['vaccines'].str.split(',').apply(len)

conCantidad.head()

fecha_limite = pd.to_datetime('2020-12-20')

antes20 = df[df['date'] < fecha_limite]
antes20.head()

pfizer = df[df['vaccines'].str.contains('Pfizer', na=False)]
pfizer.head()

with pd.ExcelWriter('resultadosReto.xlsx', engine='xlsxwriter') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)

print("Archivo resultadosReto.xlsx creado correctamente.")
```
